```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = '../perso_projet/csv/' #à changer pour vous très surement 
assignee = pd.read_csv(path+'assignee.csv', sep = ",")
assignor = pd.read_csv(path+'assignor.csv', sep = ",")
assignment = pd.read_csv(path+'assignment.csv',sep = ",")
assignment_coveyance = pd.read_csv(path+'assignment_conveyance.csv', sep = ",")
documentid = pd.read_csv(path+'documentid.csv', sep = ",")
```

```python
info_companies = pd.read_csv('./data/info_companies.csv')
info_patents = pd.read_csv('./data/info_patents.csv')
```

```python
documentid_red = documentid[documentid.grant_doc_num.isin(info_patents.patent_number.astype(str).values)]
documentid_red = documentid_red.reset_index(drop=True)
```

```python
documentid_red
```

```python
# On cree ici la liste des rf_id qui correspondent aux patents que l'on observe
rf_id_list = np.unique(documentid_red.rf_id.values)
```

```python
assignee_red = assignee[assignee.rf_id.isin(rf_id_list)].reset_index(drop=True)
assignor_red = assignor[assignor.rf_id.isin(rf_id_list)].reset_index(drop=True)
assignment_red = assignment[assignment.rf_id.isin(rf_id_list)].reset_index(drop=True)
assignment_coveyance_red = assignment_coveyance[assignment_coveyance.rf_id.isin(rf_id_list)].reset_index(drop=True)
```

```python
assignee_red
```

```python
documentid_red.to_csv('./Data/documentid_red.csv',index=False)
assignee_red.to_csv('./Data/assignee_red.csv',index=False)
assignor_red.to_csv('./Data/assignor_red.csv',index=False)
assignment_red.to_csv('./Data/assignment_red.csv',index=False)
assignment_coveyance_red.to_csv('./Data/assignment_conveyance_red.csv',index=False)
```
