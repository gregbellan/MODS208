```python
patent_id_marked = patents[patents.marked==1].patent_number.values
patent_id_unmarked = patents[patents.marked==0].patent_number.values

ref_id_marked = (documentid[documentid.grant_doc_num.isin(patent_id_marked)].rf_id.values)
ref_id_unmarked = (documentid[documentid.grant_doc_num.isin(patent_id_unmarked)].rf_id.values)
```

```python
assignment_type_marked = [assignment_conveyance[assignment_conveyance.rf_id == ref_id_marked[k]]["convey_ty"].to_numpy()[0] for k in range(len(ref_id_marked))]
assignment_type_unmarked = [assignment_conveyance[assignment_conveyance.rf_id == ref_id[k]]["convey_ty"].to_numpy()[0] for k in range(len(ref_id))]
```

```python
name_assignee_marked = [assignee[assignee.rf_id == ref_id_marked[k]]["ee_name"].to_numpy()[0] for k in range(len(ref_id_marked))]
name_assignee_unmarked = [assignee[assignee.rf_id == ref_id[k]]["ee_name"].to_numpy()[0] for k in range(len(ref_id))]
city_assignee_marked = [assignee[assignee.rf_id == ref_id_marked[k]]["ee_city"].to_numpy()[0] for k in range(len(ref_id_marked))]
city_assignee_unmarked = [assignee[assignee.rf_id == ref_id[k]]["ee_city"].to_numpy()[0] for k in range(len(ref_id))]
country_assignee_marked = [assignee[assignee.rf_id == ref_id_marked[k]]["ee_country"].to_numpy()[0] for k in range(len(ref_id_marked))]
country_assignee_unmarked = [assignee[assignee.rf_id == ref_id[k]]["ee_country"].to_numpy()[0] for k in range(len(ref_id))]
```

```python
assignment_record_date_marked = [assignment[assignment.rf_id == ref_id_marked[k]]["record_dt"].to_numpy()[0] for k in range(len(ref_id_marked))]
assignment_record_date_unmarked = [assignment[assignment.rf_id == ref_id[k]]["record_dt"].to_numpy()[0] for k in range(len(ref_id))]
assignment_last_update_marked = [assignment[assignment.rf_id == ref_id_marked[k]]["last_update_dt"].to_numpy()[0] for k in range(len(ref_id_marked))]
assignment_last_update_unmarked = [assignment[assignment.rf_id == ref_id[k]]["last_update_dt"].to_numpy()[0] for k in range(len(ref_id))]
```

```python
assignor_name_marked = [assignor[assignor.rf_id == ref_id_marked[k]]["or_name"].to_numpy()[0] for k in range(len(ref_id_marked))]
assignor_name_unmarked = [assignor[assignor.rf_id == ref_id[k]]["or_name"].to_numpy()[0] for k in range(len(ref_id))]
assignor_exec_date_marked = [assignor[assignor.rf_id == ref_id_marked[k]]["exec_dt"].to_numpy()[0] for k in range(len(ref_id_marked))]
assignor_exec_date_unmarked = [assignor[assignor.rf_id == ref_id[k]]["exec_dt"].to_numpy()[0] for k in range(len(ref_id))]
```

```python
assignment_type = np.array(assignment_type_marked + assignment_type_unmarked)
name_assignee = np.array(name_assignee_marked + name_assignee_unmarked)
city_assignee = np.array(city_assignee_marked + city_assignee_unmarked)
country_assignee = np.array(country_assignee_marked + country_assignee_unmarked)
assignment_record_date = np.array(assignment_record_date_marked + assignment_record_date_unmarked)
assignment_last_update = np.array(assignment_last_update_marked + assignment_last_update_unmarked)
assignor_name = np.array(assignor_name_marked + assignor_name_unmarked)
assignor_exec_date = np.array(assignor_exec_date_marked + assignor_exec_date_unmarked)
patent_number = np.array(patent_id_marked + patent_id_unmarked)
ref_number = np.array(ref_id_marked + ref_id)
marked = np.array([1]*len(patent_id_marked) + [0]*len(patent_id_unmarked))
```

```python
assignments = pd.DataFrame(columns = ["ref_id", "patent_id", "assignment_conveyance", "record_date", "last_update_date", "exec_date", "name_assignee", "city_assignee", "country_assignee", "assignor_name", "patent_marked"])
assignments["ref_id"] = ref_number
assignments["patent_id"] = patent_number
assignments["assignment_conveyance"] = assignment_type
assignments["record_date"] = assignment_record_date
assignments["last_update_date"] = assignment_last_update
assignments["exec_date"] = assignor_exec_date
assignments["name_assignee"] = name_assignee
assignments["city_assignee"] = city_assignee
assignments["coutnry_assignee"] = country_assignee
assignments["assignor_name"] = assignor_name
assignments["patent_marked"] = marked
```

```python
assignments.to_csv('./data/full_data_assignments.csv')
```
